```python
x=[1,2,3]
y=[2,4,6]
x=np.arange(0,5,0.1)
y=x**3
#plt.plot(x,y,color="yellow",marker="+")
plt.plot(x,y,color="red",linewidth="10")
plt.show()
```

```python
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
plt.plot(x,y,color="red",linewidth="5",label="x^3")
plt.plot(x,y2,color="black",linewidth="5",label="x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Matplotlib  Demo")
plt.text(4,60,"Text")
#plt.axis([0,5,0,5])
plt.grid()
plt.legend()
plt.show()
```

```python
#pie chart
sizes=[2,3,4,6]
labels=["A","B","C","D"]
colors=["red","yellow","green","pink"]
explode=[0.1,0,0,0]
plt.title("Pie Graph")
plt.pie(sizes,colors=colors,labels=labels,autopct="%.1f%%",explode=explode,counterclock=False,startangle=100,shadow=True)
plt.axis("equal")
plt.show()
```

```python
x=np.arange(0,10)
y=x*x
## Creating Subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y,"y-*") 
plt.show()
```

```python

#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

```python
## Bar Plot
x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

```python
## Histogram

a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a,color="orange") 
plt.title("histogram") 
plt.show()
```

```python
##BOX PLOT 
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);
```
