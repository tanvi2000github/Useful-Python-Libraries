# SEABORN
## -Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
print(tips.head())
```

```python
sns.barplot(x="sex",y="total_bill",data=tips)
```

```python
sns.countplot(x="sex",data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,hue="smoker")
```

```python
sns.violinplot(x="day",y="total_bill",data=tips)
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,split=True,hue="sex")
```

```python
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"],kde=False,bins=40,color="Red")
```

```python
sns.kdeplot(tips["total_bill"])
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex",color="yellow")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg",color="green")
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue="sex")
```

```python
tips_corr=tips.corr()
tips_corr
```

```python
sns.heatmap(tips_corr)
```

```python
flights=sns.load_dataset("flights")
flights
```

```python
flights_pivot=flights.pivot_table(index="month",columns="year",values="passengers")
flights_pivot
```

```python
sns.heatmap(flights_pivot)
```

```python
sns.heatmap(flights_pivot,cmap="cool")
```
